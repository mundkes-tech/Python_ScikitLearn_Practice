```python
import pandas as pd

data = {
    "Hours": [2, 4, 6, 8],
    "Attendance": [80, 85, 90, 95],
    "Assignments": [5, 7, 9, 10],
    "Marks": [40, 60, 80, 95]
}

df = pd.DataFrame(data)

print(df)
```

```python
x = df[["Hours","Attendance","Assignments"]]
print(x)  # Feature (X) - Input,Known

# Why is X usually 2D?
# Because Scikit-learn expects data in:
# Rows × Columns 
# format.
```

```python
y=df["Marks"]
print(y)  #Target (y) - Output,Predicted
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = {
    "Hours": [2, 4, 6, 8, 5, 7, 3, 9],
    "Attendance": [80, 85, 90, 95, 88, 92, 82, 97],
    "Marks": [40, 60, 80, 95, 70, 90, 50, 98]
}

df = pd.DataFrame(data)

x=df[["Hours","Attendance"]]
y=df["Marks"]

X_train,X_test,Y_train,Y_test = train_test_split(
    x,
    y,
    test_size = 0.2,  # means Testing data 20%
    random_state = 33 # can write anything
)

print(X_train)
print("----------------------------")
print(X_test)
print("----------------------------")
print(Y_train)
print("----------------------------")
print(Y_test)
```
